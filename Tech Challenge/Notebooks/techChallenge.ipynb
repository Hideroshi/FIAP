Pre-processamento dos dados:

```python
dados.head()
```

aplicando label encoding para as categorias com 2 tipos de categoria.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

colunas=['genero','fumante']

# Ajustar e transformar os r√≥tulos
for col in colunas:
    dados[col] = label_encoder.fit_transform(dados[col])

dados.head()
```

Aplicando Hot enconding para as demais colunas

```python
dummy_region=pd.get_dummies(dados['regiao'], prefix='dummy')

dados_hotencoded = pd.concat([dados,dummy_region],axis=1)
dados_hotencoded.drop(['regiao',],axis=1, inplace=True)
dados_hotencoded.head()
```

# Feature Scaling

```python
dados_hotencoded.head()
```

```python
dados_hotencoded.shape
```

```python
y = dados_hotencoded["encargos"]
x = dados_hotencoded.drop(columns=["encargos"], axis=1)

x.head()
```

Fun√ß√£o de treino e avalia√ß√£o dos modelos.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error

def treinar_e_avaliar(x_train, x_test, y_train, y_test, model):
  _model = model()
  _model.fit(x_train, y_train)

  predicoes = _model.predict(x_test)

  r_2 = r2_score(y_test, predicoes)
  mae = mean_absolute_error(y_test, predicoes)
  mape = mean_absolute_percentage_error(y_test, predicoes)

  print("r_2", r_2)
  print("mae", mae)
  print("mape", mape)

  return _model
```

# Modelo sem Scaler:

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)

treinar_e_avaliar(x_train, x_test, y_train, y_test, RandomForestRegressor)
treinar_e_avaliar(x_train, x_test, y_train, y_test, LinearRegression)
```

# Modelos com Scaler:

MinMax:

```python
from sklearn.preprocessing import MinMaxScaler

x_minMaxScaler = x
y_minMaxScaler = y

x_train, x_test, y_train, y_test = train_test_split(x_minMaxScaler, y_minMaxScaler, test_size=0.4, random_state=42)

scaler = MinMaxScaler() #chamando o metodo de normaliza√ß√£o dos dados (0-1)
scaler.fit(x_train) # Aqui ele calcula os parametros da formula

x_train_normalized = scaler.transform(x_train)
x_test_normalized = scaler.transform(x_test)

pd.DataFrame(x_train_normalized)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)

treinar_e_avaliar(x_train_normalized, x_test_normalized, y_train, y_test, RandomForestRegressor)
treinar_e_avaliar(x_train_normalized, x_test_normalized, y_train, y_test, LinearRegression)
```

```python
from sklearn.preprocessing import StandardScaler

x_standardScaler = x
y_standardScaler = y

X_train, X_test, y_train, y_test = train_test_split(x_standardScaler, y_standardScaler, test_size=0.2, random_state=42)

scaler = StandardScaler() #chamando o metodo de padroniza√ß√£o dos dados (m√©dia e std)
scaler.fit(x_train) # qual m√©dia e std ser√° utilizado para o escalonamento

x_train_standardized = scaler.transform(x_train)
x_test_standardized  = scaler.transform(x_test)

pd.DataFrame(x_train_standardized)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)

treinar_e_avaliar(x_train_standardized, x_test_standardized, y_train, y_test, RandomForestRegressor)
treinar_e_avaliar(x_train_standardized, x_test_standardized, y_train, y_test, LinearRegression)
```
